## Reasons to approximate functions

Imagine you have the following function of y(x):

```python
# Import libraries 
import matplotlib.pyplot as plt 
import numpy as np 

# describe function
x = np.linspace(0, 4, 50)
y = 7.33*(x**5) - 72.3*(x**4) + 253*(x**3) - 368*(x**2) + 250*x + 0.02
# Plot 
plt.plot(x, y, 'black', label ='somefunction of x')

```

If you really need to give a full picture of this function, it would be:

> $y = 7.33x^5-72.3x^4+253x^3-368x^2+250x+0.02$

and with this function you could get the exact point you need:

```python
# Plot 
plt.plot(x, y, 'black', label ='somefunction of x',marker = 'o', markeredgecolor = 'blue',markevery = [20])
```

But what if we could draw a line with the same height and slope as the function? We can do this with the taylor series, and end up with a fairly good approximation of a function way less complicated, with could be usefull in a lot of real world applications and situations:

(red line is the approximated function)

```python
z = 50*x + 15
plt.plot(x, y, 'black', label ='somefunction of x',marker = 'o', markeredgecolor = 'blue',markevery = [20])
plt.plot(x, z,'r')
plt.show
```

# Power series (Taylor series)

Power series let's us approximate functions with lot's of powers in to function with less power's. Visualize the following function:

$g(x) = a + bx + cx^2 + dx^3 + ex^4 ... + zx^n$

We could have potentially infinity powers, but what if we could just use a few power terms and have a decent approximation? That's the objective here.

The nomenclature is as follows, if we approximate our g(x) function to this:

$g(x) = a$ <- 0'th order approximation
$g(x) = a + bx$ <- 1'th order approximation
$g(x) = a + bx + cx^2$ <- 2'th order approximation, and so on.

Those are called truncated series.

# Maclaurin series

The theory is straightforward: If i know eveything from a function at point 0, i know everything from the function. 

Note that this is specific for a well-behaved function (continous and differentiable as many times as needed).

The Maclaurin series is specific to evaluating the function at the point 0, with the following formula:

# $g(x) = \sum_{n=0}^\infty \frac{f^{n}(0)}{n!}x^n$



Look at what looks like a 4'th order approximation:

# $ g_4(x) = \frac{f^{(0)}(0)}{0!} +  \frac{f^{(1)}(0)}{1!} +\frac{f^{(2)}(0)}{2!} +\frac{f^{(3)}(0)}{3!} + \frac{f^{(4)}(0)}{4!} $

note that the $f^{(n)}$ notation means the order of the derivative, so $f^{(2)}(0)$ is the second derivative of f(x) evaluated at point 0

# The taylor series (finally)

The taylor series is pretty much the same as the Maclaurin, with the difference that we no longer have the restraint of the series at point 0. If we know everything from the function at any point, we know everything from the function

To do this we just need to do a small adjustment to the Maclaurin function:


# $g(x) = \sum_{n=0}^\infty \frac{f^{n}(p)}{n!}(x-p)^n$

Now we can evaluate any given point! And if we evaluate at p = 0, our expression reduces to the maclaurin.

# Linearisation

Now we are going to understand the error in approximations.

Remember our initial formula: gradient = $\frac{rise}{run}$ 

We know that the gradient of the f(p) is f'(p) and that the run is equivalent to the distance between point p and point x in the graph (x - p), so now we have:

$f'(p) = \frac{rise}{x-p}$, and if we want to find our rise (the height x gained in respect to p) : $rise = f'(p)(x-p)$

Now we are going to call the distance between x and p (x-p) = $\Delta p$
And with this notation we can call x = $p + \Delta p$

So our first order approximation in the taylor series (remember that its a straight line?) is:

$g_1(x) = f(p) + f'(p)(x-p)$ and with this new notation for p is equal to:

$g_1(p + \Delta p) = f(p) + f'(p)(\Delta p)$

Now let's change p notation to x, just for simplicity:

$g_1(x + \Delta x) = f(x) + f'(x)(\Delta x)$

Now we know that at point x our function gives us the exact point in the function. But if we use this function (the straight line, remember? linearisation?) to evaluate other point that is not x, what would our error be?

Well, we know that if we extended this function infinitely (its a taylor series!) we would get a perfect function. And if our $\Delta x$ is a small number (small step) our $\Delta x^2$(the second order of the taylor series) is a very small number, and we could call this our error term:

$g_1(x + \Delta x) = f(x) + f'(x)(\Delta x) + O(\Delta x^2)$ where $O(\Delta x^2)$ represents our error which is second order accurate.
